```python
def apply_frequency_encoding(df, columns_to_encode):

    frequency_mappings = {}

    for column in columns_to_encode:
        category_frequencies = df[column].value_counts()
        frequency_mapping = category_frequencies.to_dict()
        df[column + '_freq_encoded'] = df[column].map(frequency_mapping)
        frequency_mappings[column] = frequency_mapping

    return df, frequency_mappings

# Example usage:
# Assuming you have a DataFrame 'df' and a list of columns to encode 'columns_to_encode'.
df, frequency_mappings = apply_frequency_encoding(df, ['cuisines','locality','rest_type'])
```
